```python
import matplotlib.pyplot as plt
import numpy as np

# ... (Your variable definitions remain here) ...
drop_out_point = [-0.05, -0.05, 0.0] 
object_A_init_pose = [0.02, 0.06, 0.0] 
object_B_init_pose = [0.05, 0.05, 0.0] 
init_pose = [0.0, 0.0, 0.98]

ee_x_index = 3 
ee_y_index = 4 

# Helper variables for arrow frequency and size
skip = 10  # Plot an arrow every 20 data points (adjust this if too crowded/sparse)

# --- Indirect path ---
plt.figure() 
x_ind = X_indirect[:, ee_x_index]
y_ind = X_indirect[:, ee_y_index]

# 1. Plot the main line
plt.plot(x_ind, y_ind, zorder=1) 

# 2. Add Arrows
# Calculate directions (u, v)
u = np.diff(x_ind)
v = np.diff(y_ind)
# Subsample the data so we don't have an arrow on every single pixel
pos_x = x_ind[:-1][::skip]
pos_y = y_ind[:-1][::skip]
dir_x = u[::skip]
dir_y = v[::skip]
# Normalize vectors to make all arrows the same size (optional)
norm = np.sqrt(dir_x**2 + dir_y**2)
# Avoid division by zero
mask = norm > 0
plt.quiver(pos_x[mask], pos_y[mask], dir_x[mask]/norm[mask], dir_y[mask]/norm[mask], 
           angles='xy', scale_units='xy', scale=200, width=0.005, color='black', zorder=2)
# Note: 'scale' is inverse size (smaller number = longer arrows). 

# 3. Plot Markers
plt.scatter(drop_out_point[0], drop_out_point[1], label='drop_out', s=150, marker='x', color='red', zorder=10)
plt.scatter(object_A_init_pose[0], object_A_init_pose[1], label='object_A', s=150, zorder=10)
plt.scatter(object_B_init_pose[0], object_B_init_pose[1], label='object_B', s=150, zorder=10)
plt.scatter(init_pose[0], init_pose[1], label='init_pose', s=150, zorder=10)

plt.xlim([-0.1, 0.1])
plt.ylim([-0.1, 0.1])
plt.title("End Effector X and Y trajectory [Indirect Path]") 
plt.xlabel("X Axis Displacement [m]") 
plt.ylabel("Y Axis Displacement [m]") 
plt.legend()
plt.show() 
plt.close() 


# --- Direct path ---
plt.figure() 
x_dir = X_direct[:, ee_x_index]
y_dir = X_direct[:, ee_y_index]

# 1. Plot the main line
plt.plot(x_dir, y_dir, zorder=1)

# 2. Add Arrows
u = np.diff(x_dir)
v = np.diff(y_dir)
pos_x = x_dir[:-1][::skip]
pos_y = y_dir[:-1][::skip]
dir_x = u[::skip]
dir_y = v[::skip]
norm = np.sqrt(dir_x**2 + dir_y**2)
mask = norm > 0
plt.quiver(pos_x[mask], pos_y[mask], dir_x[mask]/norm[mask], dir_y[mask]/norm[mask], 
           angles='xy', scale_units='xy', scale=200, width=0.005, color='black', zorder=2)

# 3. Plot Markers
plt.scatter(drop_out_point[0], drop_out_point[1], label='drop_out', s=150, marker='x', color='red', zorder=10)
plt.scatter(object_A_init_pose[0], object_A_init_pose[1], label='object_A', s=150, zorder=10)
plt.scatter(object_B_init_pose[0], object_B_init_pose[1], label='object_B', s=150, zorder=10)
plt.scatter(init_pose[0], init_pose[1], label='init_pose', s=150, zorder=10)

plt.xlim([-0.1, 0.1])
plt.ylim([-0.1, 0.1])
plt.title("End Effector X and Y trajectory [Direct Path]") 
plt.xlabel("X Axis Displacement [m]") 
plt.ylabel("Y Axis Displacement [m]") 
plt.legend()
plt.show() 
plt.close()
```
